# Memoria Analítica del Estudio de Vinos Españoles

## 1. Procesamiento y Limpieza de Datos

### 1.1 Tratamiento Inicial
- **Normalización de regiones**: Convertí todos los nombres de regiones a minúsculas para estandarización.
- **Imputación de valores faltantes**:  
  - Para la columna `type`: utilicé los valores de `region` cuando faltaban datos.  
  - Para `body` y `acidity`: calculé la media por región para rellenar los valores ausentes.  
- **Eliminación de outliers**: Eliminé filas con valores nulos que no podían ser imputados.

### 1.2 Segmentación Estratégica
- Dividí el dataset en dos grupos clave:
  - **Vinos económicos**: Precio ≤ 100€ (6,931 registros).
  - **Vinos premium**: Precio > 100€ (539 registros).
- Con esta división, llevé a cabo análisis comparativos entre segmentos de mercado.

## 2. Análisis de Hipótesis

### 2.1 Hipótesis Principal (Calidad-Precio)
- Calculé matrices de correlación para ambos segmentos.
- **Resultados**:  
  - Correlación moderada (0.546) en vinos premium.  
  - Correlación más débil en vinos económicos.
- **Visualicé con**:
  - Mapas de calor de correlaciones.  
  - Gráficos de dispersión precio-calificación.

### 2.2 Hipótesis Regionales
- Agregué datos por región:
  - Calificación media.  
  - Precio promedio.  
  - Cantidad de vinos.  
- **Patrones identificados**:  
  - **Regiones tradicionales** (La Rioja, Ribera del Duero):  
    - Media: 4.30.  
    - Mediana: 4.2.  
  - **Otras regiones**:  
    - Media: 4.37.  
    - Precio medio: 51.86€.

### 2.3 Análisis de Bodegas
- Clasifiqué bodegas por reconocimiento:
  - **Reconocidas**: Vega Sicilia, Contino.  
    - Mediana: ~4.6.  
  - **Otras bodegas**:  
    - Mediana: ~4.4.  
- **Métricas adicionales**:
  - Número de reseñas.  
  - Consistencia en calificaciones.

### 2.4 Perfil Sensorial
- Analicé la relación acidez-cuerpo:  
  - Correlación inversa (-0.45).  
  - Gráficos de dispersión.  
- **Clusters identificados**:  
  - Tintos potentes (>4.5 cuerpo/<1.25 acidez).  
  - Vinos equilibrados.

## 4. Conclusiones Validación

### 4.1 Hallazgos Principales
- **Relación calidad-precio**:  
  - Confirmada pero con matices.  
  - Más fuerte en el segmento premium.
- **Efecto región**:  
  - Las regiones tradicionales mantienen su prestigio.  
  - Otras regiones ofrecen mejor valor.
- **Bodegas**:  
  - Las reconocidas son más consistentes.  
  - Hay joyas ocultas en bodegas pequeñas.
- **Perfiles**:  
  - Patrón español (cuerpo alto/acidez baja).  
  - No determinante en calificaciones.
